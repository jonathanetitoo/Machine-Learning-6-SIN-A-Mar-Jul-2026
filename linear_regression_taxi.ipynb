## Define functions to build and train a model

The code you need to build and train your model is in the **Define ML functions** code cell. If you would like to explore this code, expand the code cell and take a look.

**Instructions**

```python
#@title Code - Define ML functions

def create_model(
    settings: ml_edu.experiment.ExperimentSettings,
    metrics: list[keras.metrics.Metric],
) -> keras.Model:
  """Create and compile a simple linear regression model."""
  # Describe the topography of the model.
  # The topography of a simple linear regression model
  # is a single node in a single layer.
  inputs = {name: keras.Input(shape=(1,), name=name) for name in settings.input_features}
  concatenated_inputs = keras.layers.Concatenate()(list(inputs.values()))
  outputs = keras.layers.Dense(units=1)(concatenated_inputs)
  model = keras.Model(inputs=inputs, outputs=outputs)

  # Compile the model topography into code that Keras can efficiently
  # execute. Configure training to minimize the model's mean squared error.
  model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=settings.learning_rate),
                loss="mean_squared_error",
                metrics=metrics)

  return model


def train_model(
    experiment_name: str,
    model: keras.Model,
    dataset: pd.DataFrame,
    label_name: str,
    settings: ml_edu.experiment.ExperimentSettings,
) -> ml_edu.experiment.Experiment:
  """Train the model by feeding it data."""

  # Feed the model the feature and the label.
  # The model will train for the specified number of epochs.
  features = {name: dataset[name].values for name in settings.input_features}
  label = dataset[label_name].values
  history = model.fit(x=features,
                      y=label,
                      batch_size=settings.batch_size,
                      epochs=settings.number_epochs)

  return ml_edu.experiment.Experiment(
      name=experiment_name,
      settings=settings,
      model=model,
      epochs=history.epoch,
      metrics_history=pd.DataFrame(history.history),
  )

print("SUCCESS: defining linear regression functions complete.")
```

# Part 5 - Análisis de Underfitting y Overfitting

---

En esta sección se modifican los hiperparámetros de entrenamiento para demostrar
dos escenarios clásicos de Machine Learning:

- **Underfitting**: El modelo NO logra aprender los patrones subyacentes de los datos.
- **Overfitting**: El modelo MEMORIZA los datos de entrenamiento y pierde capacidad de generalización.

### Hiperparámetros originales (referencia)
| Parámetro | Valor original |
|---|---|
| `learning_rate` | 0.001 |
| `number_epochs` | 20 |
| `batch_size` | 50 |

### Estrategia de modificación
Se ajustan `learning_rate`, `number_epochs` y `batch_size` con valores **moderados**
(no extremos) para provocar cada escenario de forma realista.

## 5.1 — UNDERFITTING (Subajuste)

### ¿Por qué estos hiperparámetros causan underfitting?

| Hiperparámetro | Valor | Efecto |
|---|---|---|
| `learning_rate` | **0.0001** | Pasos de actualización muy pequeños → el modelo avanza muy lentamente hacia el mínimo de la función de pérdida |
| `number_epochs` | **3** | Muy pocas pasadas sobre los datos → el modelo no tiene tiempo suficiente para converger |
| `batch_size` | **512** | Lotes grandes → gradientes más suavizados, menos actualizaciones por época → convergencia más lenta |

**Resultado esperado**: RMSE alto y que NO converge. La curva de pérdida se mantiene elevada.
El modelo produce predicciones pobres porque no ha aprendido la relación entre features y label.

> **Nota**: Los valores NO son extremos (el learning_rate no es 1e-8, las epochs no son 1,
> el batch_size no es el dataset completo), pero la *combinación* de los tres impide que el modelo aprenda.

### Experimento 1 — Underfitting con una feature (TRIP_MILES)

```python
#@title UNDERFITTING - Experimento 1 (una feature: TRIP_MILES)

# ============================================================
# UNDERFITTING: Hiperparámetros que impiden el aprendizaje
# - learning_rate bajo (0.0001): pasos de gradiente muy pequeños
# - epochs reducidas (3): muy pocas iteraciones sobre los datos
# - batch_size grande (512): menos actualizaciones por época
# ============================================================

settings_underfit_1 = ml_edu.experiment.ExperimentSettings(
    learning_rate = 0.0001,   # 10x menor que el original (0.001)
    number_epochs = 3,        # 6.7x menor que el original (20)
    batch_size = 512,         # 10x mayor que el original (50)
    input_features = ['TRIP_MILES']
)

metrics_underfit_1 = [keras.metrics.RootMeanSquaredError(name='rmse')]
model_underfit_1 = create_model(settings_underfit_1, metrics_underfit_1)

experiment_underfit_1 = train_model(
    'underfit_one_feature',
    model_underfit_1,
    training_df,
    'FARE',
    settings_underfit_1
)

print('\n--- UNDERFITTING - Experimento 1 (TRIP_MILES) ---')
print(f'RMSE final: {experiment_underfit_1.metrics_history["rmse"].iloc[-1]:.4f}')
print('Nota: RMSE alto indica que el modelo NO aprendió los patrones.\n')

ml_edu.results.plot_experiment_metrics(experiment_underfit_1, ['rmse'])
ml_edu.results.plot_model_predictions(experiment_underfit_1, training_df, 'FARE')
```

### Experimento 2 — Underfitting con una feature (variación de hiperparámetros)

```python
#@title UNDERFITTING - Experimento 2 (una feature: TRIP_MILES - variación)

# ============================================================
# UNDERFITTING: Segunda combinación de hiperparámetros
# - learning_rate moderadamente bajo (0.0005)
# - epochs muy pocas (2)
# - batch_size grande (256)
# ============================================================

settings_underfit_2 = ml_edu.experiment.ExperimentSettings(
    learning_rate = 0.0005,   # 2x menor que el original
    number_epochs = 2,        # 10x menor que el original
    batch_size = 256,         # 5x mayor que el original
    input_features = ['TRIP_MILES']
)

metrics_underfit_2 = [keras.metrics.RootMeanSquaredError(name='rmse')]
model_underfit_2 = create_model(settings_underfit_2, metrics_underfit_2)

experiment_underfit_2 = train_model(
    'underfit_one_feature_v2',
    model_underfit_2,
    training_df,
    'FARE',
    settings_underfit_2
)

print('\n--- UNDERFITTING - Experimento 2 (TRIP_MILES variación) ---')
print(f'RMSE final: {experiment_underfit_2.metrics_history["rmse"].iloc[-1]:.4f}')
print('Nota: RMSE alto confirma el underfitting.\n')

ml_edu.results.plot_experiment_metrics(experiment_underfit_2, ['rmse'])
ml_edu.results.plot_model_predictions(experiment_underfit_2, training_df, 'FARE')
```

### Experimento 3 — Underfitting con dos features (TRIP_MILES + TRIP_MINUTES)

```python
#@title UNDERFITTING - Experimento 3 (dos features: TRIP_MILES + TRIP_MINUTES)

# ============================================================
# UNDERFITTING con dos features
# Misma estrategia: learning_rate bajo + pocas epochs + batch grande
# Incluso con más información (2 features), el modelo no converge
# ============================================================

settings_underfit_3 = ml_edu.experiment.ExperimentSettings(
    learning_rate = 0.0001,   # 10x menor que el original
    number_epochs = 3,        # 6.7x menor que el original
    batch_size = 512,         # 10x mayor que el original
    input_features = ['TRIP_MILES', 'TRIP_MINUTES']
)

metrics_underfit_3 = [keras.metrics.RootMeanSquaredError(name='rmse')]
model_underfit_3 = create_model(settings_underfit_3, metrics_underfit_3)

experiment_underfit_3 = train_model(
    'underfit_two_features',
    model_underfit_3,
    training_df,
    'FARE',
    settings_underfit_3
)

print('\n--- UNDERFITTING - Experimento 3 (TRIP_MILES + TRIP_MINUTES) ---')
print(f'RMSE final: {experiment_underfit_3.metrics_history["rmse"].iloc[-1]:.4f}')
print('Nota: Incluso con 2 features, el modelo no puede aprender con estos hiperparámetros.\n')

ml_edu.results.plot_experiment_metrics(experiment_underfit_3, ['rmse'])
ml_edu.results.plot_model_predictions(experiment_underfit_3, training_df, 'FARE')
```

## 5.2 — OVERFITTING (Sobreajuste)

### ¿Por qué estos hiperparámetros causan overfitting?

| Hiperparámetro | Valor | Efecto |
|---|---|---|
| `learning_rate` | **0.01** | Pasos de actualización grandes → el modelo se ajusta agresivamente a cada lote de datos |
| `number_epochs` | **80** | Muchas pasadas sobre los datos → el modelo tiene tiempo de sobra para memorizar patrones específicos (incluyendo ruido) |
| `batch_size` | **5** | Lotes muy pequeños → gradientes ruidosos que capturan patrones individuales de cada mini-lote en vez de la tendencia general |

**Resultado esperado**: El loss de entrenamiento baja mucho (el modelo memoriza),
pero si se evaluara en datos nuevos, el error sería significativamente mayor.
La curva de pérdida puede mostrar oscilaciones debido al batch_size pequeño.

### Estrategia adicional para evidenciar overfitting
Para hacer el overfitting más evidente, también entrenamos con un **subconjunto
pequeño** de los datos (500 muestras). Esto facilita la memorización.

> **Nota**: Los valores NO son extremos (learning_rate no es 1.0, epochs no son 10000,
> batch_size no es 1), pero la *combinación* de los tres fomenta la memorización.

```python
#@title Preparar subconjunto pequeño de datos para demostrar overfitting

# Usamos un subconjunto pequeño (500 muestras) para facilitar la memorización
# Con menos datos, el modelo puede memorizar los ejemplos más fácilmente
overfit_df = training_df.sample(n=500, random_state=42).copy()
overfit_df.reset_index(drop=True, inplace=True)

print(f'Subconjunto para overfitting: {len(overfit_df)} muestras (de {len(training_df)} totales)')
print('Con pocos datos + muchas epochs + batch pequeño → el modelo memoriza.')
```

### Experimento 1 — Overfitting con una feature (TRIP_MILES)

```python
#@title OVERFITTING - Experimento 1 (una feature: TRIP_MILES)

# ============================================================
# OVERFITTING: Hiperparámetros que fomentan la memorización
# - learning_rate alto (0.01): actualizaciones agresivas
# - epochs altas (80): muchas pasadas sobre pocos datos
# - batch_size pequeño (5): gradientes ruidosos, memorización
# - Subconjunto pequeño de datos (500 muestras)
# ============================================================

settings_overfit_1 = ml_edu.experiment.ExperimentSettings(
    learning_rate = 0.01,     # 10x mayor que el original (0.001)
    number_epochs = 80,       # 4x mayor que el original (20)
    batch_size = 5,           # 10x menor que el original (50)
    input_features = ['TRIP_MILES']
)

metrics_overfit_1 = [keras.metrics.RootMeanSquaredError(name='rmse')]
model_overfit_1 = create_model(settings_overfit_1, metrics_overfit_1)

experiment_overfit_1 = train_model(
    'overfit_one_feature',
    model_overfit_1,
    overfit_df,             # <-- Subconjunto pequeño
    'FARE',
    settings_overfit_1
)

print('\n--- OVERFITTING - Experimento 1 (TRIP_MILES) ---')
print(f'RMSE en datos de entrenamiento: {experiment_overfit_1.metrics_history["rmse"].iloc[-1]:.4f}')
print('Nota: RMSE bajo en entrenamiento, pero el modelo ha memorizado estos 500 ejemplos.\n')

ml_edu.results.plot_experiment_metrics(experiment_overfit_1, ['rmse'])

# Visualizar predicciones en datos de ENTRENAMIENTO (overfit_df)
ml_edu.results.plot_model_predictions(experiment_overfit_1, overfit_df, 'FARE')
```

### Evaluación: Comparar rendimiento en datos de entrenamiento vs datos completos

La clave del overfitting es que el modelo funciona bien en los datos que vio,
pero mal en datos nuevos. Aquí comparamos el RMSE.

```python
#@title Comparar RMSE: datos de entrenamiento (500) vs dataset completo

# Evaluar en el subconjunto de entrenamiento (500 muestras)
train_features = {name: overfit_df[name].values for name in settings_overfit_1.input_features}
train_eval = model_overfit_1.evaluate(train_features, overfit_df['FARE'].values, verbose=0)

# Evaluar en el dataset completo (~31,694 muestras)
full_features = {name: training_df[name].values for name in settings_overfit_1.input_features}
full_eval = model_overfit_1.evaluate(full_features, training_df['FARE'].values, verbose=0)

print('=== EVIDENCIA DE OVERFITTING ===')
print(f'RMSE en datos de entrenamiento (500 muestras):  {train_eval[1]:.4f}')
print(f'RMSE en dataset completo (31,694 muestras):     {full_eval[1]:.4f}')
print(f'\nDiferencia (brecha de generalización):          {abs(full_eval[1] - train_eval[1]):.4f}')
print('\nSi la brecha es grande → el modelo memorizó en vez de generalizar = OVERFITTING')
```

```python
#@title Code - Define functions to make predictions
def format_currency(x):
  return "${:.2f}".format(x)

def build_batch(df, batch_size):
  batch = df.sample(n=batch_size).copy()
  batch.set_index(np.arange(batch_size), inplace=True)
  return batch

def predict_fare(model, df, features, label, batch_size=50):
  batch = build_batch(df, batch_size)
  predicted_values = model.predict_on_batch(x={name: batch[name].values for name in features})

  data = {"PREDICTED_FARE": [], "OBSERVED_FARE": [], "L1_LOSS": [],
          features[0]: [], features[1]: []}
  for i in range(batch_size):
    predicted = predicted_values[i][0]
    observed = batch.at[i, label]
    data["PREDICTED_FARE"].append(format_currency(predicted))
    data["OBSERVED_FARE"].append(format_currency(observed))
    data["L1_LOSS"].append(format_currency(abs(observed - predicted)))
    data[features[0]].append(batch.at[i, features[0]])
    data[features[1]].append("{:.2f}".format(batch.at[i, features[1]]))

  output_df = pd.DataFrame(data)
  return output_df

def show_predictions(output):
  header = "-" * 80
  banner = header + "\n" + "|" + "PREDICTIONS".center(78) + "|" + "\n" + header
  print(banner)
  print(output)
  return
```

```python
#@title Code - Make predictions

output = predict_fare(experiment_3.model, training_df, experiment_3.settings.input_features, 'FARE')
show_predictions(output)
```
